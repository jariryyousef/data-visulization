```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

# resource
#https://www.youtube.com/playlist?list=PLtPIclEQf-3cG31dxSMZ8KTcDG7zYng1j
```

## Load data set

```python
# df_Restaurant=pd.read_csv('Independence100.csv')
# df_Restaurant
car_crashes = sns.load_dataset("car_crashes")
car_crashes.head(10)
```

## Density of accidents due to speed

```python
sns.distplot(car_crashes.speeding ,color='red')
```

## Proportion the accidents

```python
res_head=car_crashes
res_head
sns.ecdfplot(res_head)
```

## Total of Accidents due to alcohol

```python
graph = sns.barplot(x="alcohol", y="total", data= car_crashes)
graph
```

## Accidents no previous

```python

sns.regplot(x='no_previous', y='total', data=car_crashes, color='red')
```

## Conclusion
- The intensity of accidents due to speed decreases
- Highest accident proportion for ins premium
- The intensity of accidents due to alcohol increases
- Accidents without precedent are increasing
